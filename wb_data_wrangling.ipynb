```python
import pandas as pd 
import wbdata
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
countries = ['BFA', 'COD', 'ETH', 'KEN', 'NGA', 'SEN', 'ZAF', 'TZA', 'UGA','ZMB']
```

```python
# generate a dict from the indicators file
wd_indicator = pd.read_csv('Takwimu_indicators.csv',index_col=0, squeeze=True).to_dict()

# wd_indicator
```

```python
df = wbdata.get_dataframe(wd_indicator, country=countries, convert_date=False)
df
```

```python
# df.index
```

```python
# structure the index hierarchically by country
df.reset_index(inplace=True)
df['date'] = df['date'].astype(int)
df.set_index(['country', 'date'], inplace=True)
```

```python
df.head()
```

```python
df.index.levels
```

```python
# Access a country's indicator values 

df.loc['Ethiopia'].head()
```

```python
ug_df = df.loc['Uganda']
```

```python
ug_df[['Population Male','PopulationFemale',
       'Life expectancy-Male','Life expectancy-Female']].plot(figsize=(12, 12), subplots=True)
plt.legend(loc='best')
```

```python
# Compare countries along their GDP growth

df_GDP = df.reset_index()
df_GDP = df_GDP.pivot(index='date', columns='country', values='GDP Growth')
df_GDP.head()
```

```python
df_GDP.plot(figsize=(16, 9), title='GDP Growth of Selected African Countries');

plt.ticklabel_format(style='plain', axis='y')
```

```python
# Compare select 3 countries

df_GDP[['Ethiopia','Nigeria','South Africa']].plot(figsize=(16, 9),
                                                   title='GDP Growth of Three African Countries');
```
